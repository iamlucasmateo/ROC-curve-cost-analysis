# ROC Curve as optimization constraint in cost analysis

In this notebook we will develop a short economic model for choosing the best threshold parameter in a classification problem, in order to answer the question: which is the optimal point in a ROC curve? To do this, we will be looking not only at a Machine Learning model to use in the classification task, but we will also be analysing the task from the point of view of (economic) value. Indeed, the classification errors that ML models almost inevitably have carry economic consequences. False alarms (that is, Type I error or false positives) will lead to a waste of resources; misses (i.e., Type II errors or false negatives) will bring about an opportunity cost. 

We'll start by working with a dataset and building a model for classification.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.datasets import make_classification
```

We will create a synthetic dataset for a binary classification task, with a relatively balanced split between classes (40/60). Noisy components appear in the form of uniformative and redundant features, a low class separation, and the introduction of random assignments to the target variable y (with the flip_y) parameter.

```python
X, y = make_classification(n_samples=10000,n_features=5, n_informative=4,flip_y=0.3,
                                     n_redundant=1, n_clusters_per_class=1,
                                     class_sep=0.2, weights = [0.4], random_state=40)
```

```python
# Create pandas dataframe from dataset
df = pd.DataFrame()
for i in range(X.shape[1]):
    df['x' + str(i)] = X[:,i]
df['y'] = y
df = df.reindex(np.random.permutation(df.index)) 
df.head()
```

```python
# Take a look at the pairplot
sns.pairplot(data = df, vars = df.columns[:-1], hue = 'y')
```

```python
# Import libraries that we will be using
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve, auc
```

```python
# Split train and test sets 
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=50)
```

We will use a Random Forest classifier for this task.

```python
rfc = RandomForestClassifier(n_estimators=20)
rfc.fit(X_train, y_train)
y_pred = rfc.predict_proba(X_test)
y_pred = y_pred[:,1]
fpr,tpr,thr = roc_curve(y_test,y_pred)
```

We will create a functions that allows us to plot the ROC curve.

```python
def plot_roc_curve(fpr, tpr):
    roc_auc = auc(fpr,tpr)
    plt.figure()
    lw = 3
    plt.plot(fpr, tpr, color='darkorange',
             lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()
```

We now plot the usual ROC curve for our case.

```python
plot_roc_curve(fpr,tpr)
```

## Misclassification cost

In order to compute the optimal threshold, we will use the False Negative Rate (fnr), which is the complement of the True Positive Rate (tpr). This leaves us with two scores in the form of a decision trade-off: we can choose to have a lower False Positive Rate, at the cost of a higher False Negative Rate. This is the function that we will be using as an optimization constraint for the minimization of classification costs.

Cost ($C$) is a random variable with two possible values: $c_p$ (cost of False Positives) and $c_n$ (cost of False Negatives). These two values have a respective probability of ocurrence, which depends on the threshold ($T$) of the classification model, so we will define these conditional probabilities as $P(c_p/T=t)$ and $P(c_n/T=t)$. So the conditional expectation of misclassification Cost will be given by: $E[C/T = t] = P(c_p / T = t)*c_p + P(c_n / T = t)*c_n$. This is the expression that we would like to minimize. Let's start by developing an example.

```python
# Example
fnr = 1-tpr # False Negative Rate
cp,cn = 5,5 # cost of misclassification (false positives, false negatives)
exp = fpr*cp + fnr*cn
exp
```

```python
# Threshold that minimizes expected cost
min_ind = np.argmin(exp)
min_thr = thr[min_ind]
# Let's plot this
plt.figure()
plt.plot(thr, exp, color='darkorange',lw=3,label=f'Exp. Cost')
plt.plot(thr, fpr, color='blue',lw=3,label=f'FPR')
plt.plot(thr, fnr, color='red',lw=3,label=f'FNR')
plt.xlim([-0.01, 1.01])
# plt.ylim([-0.01, 1.05])
plt.xlabel('Threshold')
plt.ylabel('Expected Cost')
plt.title('FP Cost:{}, FN Cost: {}, Min Thr: {}'.format(5,5,min_thr))
plt.legend(loc="upper right")
plt.show()
```

By varying the cost parameters we can do some comparative statics; i.e., how does the optimal threshold change when the costs change? Because it is <b>relative</b> costs that matter in this context, we will leave $c_p$ fixed and change $c_n$. We will plot the expected cost curve (as a function of the threshold), with the corresponding ROC curve and its optimal point below.

```python
fig,ax = plt.subplots(2,4,figsize=(16,8))
colors = ['darkorange','blue','red','green']
fontdict = {'fontsize':13,'fontweight':'bold'}
cost_list = [3,5,10,50]
for i in range(4):
    cp,cn = 10,cost_list[i]
    exp = fpr*cp + fnr*cn
    min_ind = np.argmin(exp)
    min_thr = thr[min_ind]
    ax[0,i].plot(thr, exp, color=colors[i],lw=3,label=f'Exp. Cost')
    ax[0,i].scatter(thr[min_ind],exp[min_ind],color=colors[i],lw=10)
    ax[0,i].set_xlim([-0.01, 1.01])
    ax[0,i].set_xlabel('Threshold')
    ax[0,i].set_ylabel('Expected Cost')
    ax[0,i].set_title('FP Cost:{}, FN Cost: {}, Min Thr: {}'.format(cp,cn,min_thr),fontdict=fontdict)
    ax[1,i].plot(fpr, tpr, color=colors[i],lw=3)
    ax[1,i].scatter(fpr[min_ind],tpr[min_ind],color=colors[i],lw=10)
    ax[1,i].set_xlim([-0.01, 1.01])
    ax[1,i].set_xlim([0.0, 1.05])
    ax[1,i].set_xlabel('False Positive Rate')
    ax[1,i].set_ylabel('True Positive Rate')
    ax[1,i].set_title('ROC Curve',fontdict=fontdict)
plt.tight_layout()
```

Notice how the threshold moves towards lower values as the relative cost of the false negatives ($c_n$) increases. This is exactly what we would expect: the more expensive it is to miss a prediction, the more the threshold should be nudged towards positive prediction. We do this by increasing the threshold. In terms of the ROC curve, this means moving <b> upwards and rightwards</b>, therefore increasing both the True Positives and the False Positives.  

```python
# Export one picture as example
plt.figure(figsize=(5,4))
colors = ['darkorange','blue','red','green']
fontdict = {'fontsize':13,'fontweight':'bold'}
cp,cn = 10,10
exp = fpr*cp + fnr*cn
min_ind = np.argmin(exp)
min_thr = thr[min_ind]
plt.plot(thr, exp, color=colors[i],lw=3,label=f'Exp. Cost')
plt.scatter(thr[min_ind],exp[min_ind],color=colors[i],lw=10)
plt.xlim([-0.01, 1.01])
plt.xlabel('Threshold')
plt.ylabel('Expected Cost')
plt.title('FP Cost:{}, FN Cost: {}, Min Thr: {}'.format(cp,cn,min_thr),fontdict=fontdict)
plt.savefig('EconomicAnalysisExample.png')
plt.tight_layout()
```
